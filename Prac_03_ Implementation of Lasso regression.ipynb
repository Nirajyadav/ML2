```python
from sklearn.datasets import load_diabetes
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = load_diabetes()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df
```

```python
df['Target'] = data.target\


# calling target column
```

```python
df
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test  = train_test_split(data.data, data.target, test_size=0.2, random_state=2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import Ridge

from sklearn.metrics import r2_score
```

```python
coef = []

r2_scores = []

for i in [0,10,100,1000]:
    reg = Ridge(alpha=i)
    
    reg.fit(x_train,y_train)
    
    coef.append(reg.coef_.tolist())
    
    y_pred = reg.predict(x_test)
    
    r2_scores.append(r2_score(y_test, y_pred))
```

**We have 10 independent columns , we get 10 coef for each value.**

```python
coef
```

```python
plt.figure(figsize=(14,9))

plt.subplot(221)

plt.bar(data.feature_names, coef[0])

plt.title('Alpha = 0, r2_score = {}'.format(round(r2_scores[0],2)))

# --------------

plt.subplot(222)

plt.bar(data.feature_names, coef[1])

plt.title('Alpha = 10, r2_score = {}'.format(round(r2_scores[1],2)))

# --------------

plt.subplot(223)

plt.bar(data.feature_names, coef[2])

plt.title('Alpha = 100, r2_score = {}'.format(round(r2_scores[2],2)))

# ----------------

plt.subplot(224)

plt.bar(data.feature_names, coef[3])

plt.title('Alpha = 1000, r2_score = {}'.format(round(r2_scores[3],2)))
```
